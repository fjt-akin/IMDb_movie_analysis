```python
# Function to parse duration string and convert to timedelta
def parse_duration(duration_str):
    try:
        # Check if value is a string
        if isinstance(duration_str, str):
            # Split duration string into hours and minutes
            parts = duration_str.split(' ')
            hours = 0
            minutes = 0
            for part in parts:
                if 'h' in part:
                    hours = int(part.replace('h', ''))
                elif 'm' in part:
                    minutes = int(part.replace('m', ''))
            
            # If only hours are provided, set minutes to 0
            if hours and not minutes:
                minutes = 0
            
            return pd.Timedelta(hours=hours, minutes=minutes)
        else:
            return pd.NaT  # Return NaT for non-string values
    except ValueError:
        return pd.NaT  # Return NaT for invalid durations

# Convert 'duration' column to timedelta
movie_df['Duration'] = movie_df['Duration'].apply(parse_duration)
```

<h4> 2) - Are certain genres more prevalent in different decades</h4>

```python
#define a function to get decade
def get_decade(year):
    return int((year.year // 10) * 10)

#create a new column called Decade to represent the decade each movie belongs to
movie_df['Decade'] = movie_df['Year'].apply(get_decade)
movie_df.head(5)
```

```python
#split genres and create a list of all unique genres
all_genres = movie_df['Genre'].str.split(',').sum()
stripped_genres = [genre.strip() for genre in all_genres]
unique_genre = set(stripped_genres)
unique_genre

```

```python
#initializes an empty dictionary genre_counts_per_decade to store genre counts for each decade.
genre_counts_per_decade = {}
#iterates over each decade and the corresponding group of movies in the DataFrame movie_df grouped by the 'Decade' column
for decade, group in movie_df.groupby('Decade'):
    #initialize an empty dictionary inside genre_counts_per_decade for the current decade to store genre counts.
    genre_counts_per_decade[decade] = {}
    #concatenates the genres of all movies in the current decade into a single list decade_genres.
    decade_genres = group['Genre'].str.split(', ').sum()
    #This loop counts the occurrences of each unique genre in the decade_genres list and stores the counts in the genre_counts_per_decade dictionary for the current decade.
    for genre in unique_genre:
        genre_counts_per_decade[decade][genre] = decade_genres.count(genre)
```

```python
#create a transposed DataFrame genre_counts_df from the genre_counts_per_decade dictionary for easy plotting, where each row represents a decade and each column represents a genre.
genre_counts_df = pd.DataFrame(genre_counts_per_decade)
data = genre_counts_df.T
data
```

```python
#visualize to give us a stacked bar plot using the DataFrame data, where each bar represents a decade, 
# and the height of each segment within the bar represents the number of movies belonging to a particular 
# genre in that decade.

data.plot(kind='bar',stacked=True, figsize=(15,7))

plt.title('Prevalence of Genres across Decades')
plt.xlabel('Decade')
plt.ylabel('Number of Movies')
plt.legend(title='Genre')
plt.xticks(rotation=45)
plt.show()
```

Concentrating on early 2000's to 2023, Drama seems to be a popular movie genre, followed closely by action, then comedy.

<h4>3) - Is there a correlation between popularity of a genre and its Metascore ?</h4>

<p>What is a metascore ? 

A Metascore is a weighted average score assigned to a movie, video game, music album, or other forms of media based on reviews from professional critics. It is generated by the website Metacritic, which aggregates reviews from various reputable sources and assigns scores to each review. The Metascore provides a quantitative measure of the critical consensus on the quality of the media product.

For movies, the Metascore typically ranges from 0 to 100, with higher scores indicating generally favorable reviews and lower scores indicating generally unfavorable reviews. Similarly, for video games and music albums, Metascores range from 0 to 100, serving as a helpful tool for consumers to gauge the overall critical reception of a particular work.</p>

```python
#using votes to measure popularity of a genre
popularity_by_genre = genre_df.groupby('Genre')['Votes'].mean()
#Calculate average Metascore
average_metascore_per_genre = genre_df.groupby('Genre')['Meta Score'].mean()
```

```python
#plot the results
plt.figure(figsize=(10, 6))
plt.scatter(popularity_by_genre,average_metascore_per_genre)
plt.title('Popularity vs Average Metascore per genre')
plt.xlabel('Average number of votes')
plt.ylabel('Avergae Metascore')
plt.grid(True)
plt.show()
```

```python
#calculate the correlation coefficient
correlation = popularity_by_genre.corr(average_metascore_per_genre)
print('correlation coefficient:', correlation)
```

<ul>
<li>If 'r' is close to 1, it indicates a strong positive correlation, meaning that as one variable increases, the other variable also tends to increase.</li>
<li>If 'r' is close to -1, it indicates a strong negative correlation, meaning that as one variable increases, the other variable tends to decrease.</li>
<li>If 'r' is close to 0, it indicates a weak or no correlation between the variables.</li>
</ul>



<p> from the scatter plot above and correlation coefficient, it shows there is no correlation between the popularity and metascore per genre</p>

<h4>2) - Do movies with higher ratings tend to have more votes ?</h4>

```python
#plot the results
plt.figure(figsize=(10, 6))
plt.scatter(average_rating_per_genre,popularity_by_genre)
plt.title('Relationship between movie ratings and votes')
plt.xlabel('Average number of ratings')
plt.ylabel('Avergae number of votes')
plt.grid(True)
plt.show()
```

```python
#calculating the Pearson correlation coefficient to quantify the strength and direction of this relationship.

from scipy.stats import pearsonr
# Calculate Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(genre_df['Rating'], genre_df['Votes'])

# Print correlation coefficient
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.2f}")

# Interpret the correlation coefficient
if correlation_coefficient > 0.5:
    correlation_strength = 'strong positive'
elif correlation_coefficient < -0.5:
    correlation_strength = 'strong negative'
elif correlation_coefficient > 0.3:
    correlation_strength = 'moderate positive'
elif correlation_coefficient < -0.3:
    correlation_strength = 'moderate negative'
else:
    correlation_strength = 'weak or no'

print(f"Correlation Strength: {correlation_strength}")
print(f"P-value: {p_value:.4f}")
```
